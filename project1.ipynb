# gender distribution

```python
ax=sns.countplot(x="Gender",data=df,hue="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# from the above chart we analysed that the no of females in the data than the number of boys

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb, annot=True)
plt.title("realtionship b/w parents education and Student's score")
plt.show()
```

# from the above chart we have concluded that the education of the parents have a good impact.

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1, annot=True)
plt.title("Relationship b/w parent's education and student's score")
plt.xticks(rotation="vertical")
plt.show()
```

# From the above chart we have concluded that marital status has no impact on children education.
